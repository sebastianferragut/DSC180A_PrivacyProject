```python
import pandas as pd
import numpy as np
import json
from pprint import pprint
```

```python
with open('www.linkedin.com_crawl_results.json') as fd:
     json_data = json.load(fd)
     pprint(json_data)
```

```python
visited_links = json_data["visited_links"]
```

```python
layer_dict = json_data['layer_dict']
layer_dict
```

```python
sum_ = 0

for key in json_data['layer_dict']:
    sum_ += layer_dict[key]
    layer_dict[key] = sum_

layer_dict
```

```python
new_dict = {}

for key in layer_dict:
    if key == "Layer 0":
        new_dict[key] = visited_links[ : layer_dict[key]]
    elif key == "Layer 7":
        new_dict[key] = visited_links[layer_dict[key] : ]
    else:
        start = int(layer_dict['Layer ' + str(int(key.split(" ")[1]) - 1)])
        end = int(layer_dict[key])

        print(start, end)
        
        new_dict[key] = visited_links[start : end]
```
